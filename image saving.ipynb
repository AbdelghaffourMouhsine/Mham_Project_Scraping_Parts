```python
import requests
import os

# Créer le dossier 'photos' s'il n'existe pas
if not os.path.exists('photos'):
    os.makedirs('photos')
    
img_data = requests.get("https://www.rockauto.com/info/28/10268-000__ra_m.jpg").content
with open(f'photos/image_2.jpg', 'wb') as handler:
    handler.write(img_data)
```

```python
import cv2

# Lire l'image
image = cv2.imread('photos/image_2.jpg')

# Afficher l'image
cv2.imshow('Image', image)

# Attendre une touche pour fermer la fenêtre
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Lire l'image
image = cv2.imread('photos/image_2.jpg')

# Convertir l'image de BGR à RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Afficher l'image avec Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Pour masquer les axes
plt.show()
```
